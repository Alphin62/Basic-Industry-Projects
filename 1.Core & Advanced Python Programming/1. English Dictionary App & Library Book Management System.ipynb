### 1A: English Dictionary App

### This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app that returns definitions of English words.

```python
import json
print('Main Menu \n\n')
print('1. Add a new word \n2. Find the meaning \n3. Update a word \n4. Exit \n')
dict1={}
while True :
    try:
        choice=int(input('Enter the choice : '))
    except ValueError:
        print('************    ENTER ONLY NUMBERS     ************')
        
    else:
        if choice == 1:
            word=input('Enter the word : ')
            meaning=input('Enter the meaning : ')
            dict1[word]=meaning
            with open ('word.txt','w') as f:
                json.dump(dict1,f)
                
            print('**********    Word added successfully    **********\n\n')
            
        elif choice == 2:
            word=input('Enter the word : ')
            #find_meaning=input('Enter the word : ')
            with open ('word.txt','r+') as f:
                data=json.load(f)
                if word in data.keys():
                    for key,value in data.items():
                        if key==word:
                            print('Meaning of {} : {}'.format(key,value))
                              
                else:
                    print('**********    Word is not available    **********\n\n')
        
        elif choice == 3:
            word=input('Enter the word : ')
            new_meaning=input('Enter new meaning : ')
            f=open ('word.txt','r')
            hi=json.load(f)
            f.close()
            if word in hi:
                hi[word]=new_meaning
                f=open('word.txt','w')
                json.dump(hi,f)
                f.close()
                print('**********    Word updated successfully    **********\n\n')
                    
            else:
                print('**********    Word is not available    **********\n\n')
                    
                                   
        elif choice == 4:
            exit=input('Are you sure you want to exit ? (yes/no)')
            if exit == 'y' or exit == 'yes' or exit == 'Yes' or exit =='YES':
                print('**********    Bye Bye    **********\n\n')
                break
            else:
                continue
        

        else:
            print('**********    Enter only the given choices    **********\n\n')


```

### 1B: Library Book Management System

### In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement CRUD( create, read, update, delete) operations on a ‘books’ database, in SQLite. The window will contain text widgets to accept / display book title, author, year of publication, ISBN (optional) and a large text area/ list box to display the records which match the search criteria.

```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as mb

root= Tk()
root.title('A Library Application!')

def Database():
    global con, cur
    con=sqlite3.connect('simple.db')
    cur=con.cursor()
    cur.execute('CREATE TABLE IF NOT EXISTS Librarybooks(Booktitle TEXT,Author TEXT,Year TEXT,ISBN TEXT)')
    
def Create():
    if BOOKTITLE.get() == "" or AUTHOR.get() == "" or YEAR.get() == "" or ISBN.get() == "" :
        txt_result.config(text='Please enter all the fields',fg='red')
        
    else:
        Database()
        cur.execute('INSERT INTO Librarybooks(Booktitle,Author,Year,ISBN) VALUES (?,?,?,?)',(str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())))
        con.commit()
        
        BOOKTITLE.set("")
        AUTHOR.set("")
        YEAR.set("")
        ISBN.set("")
        cur.close()
        con.close()
        txt_result.config(text='Record inserted !!!!!!!',fg='green')
        
def Read():
    tree.delete(*tree.get_children())
    Database()
    cur.execute('SELECT * FROM Librarybooks ORDER BY Author ASC')
    fetch=cur.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
        cur.close()
        txt_result.config(text='Data fetched !!!!!',fg='green')

def Search():
    if BOOKTITLE.get() == "" and AUTHOR.get() == "":
        txt_result.config(text='Kindly enter the data',fg='cyan')
    else:
        Database()
        tree.delete(*tree.get_children())
        cur.execute("SELECT * FROM Librarybooks where Booktitle =? or Author = ? ",(str(BOOKTITLE.get()),(str(AUTHOR.get()))))
        rows=cur.fetchall()
        for data in rows:
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            cur.close()
            txt_result.config(text='Data found !!!!!!',fg='blue')
    
def Delete():
    if BOOKTITLE.get() == "" :
        txt_result.config(text='Kindly enter the BOOKTITLE',fg='cyan')
    else:
        Database()
        cur.execute("DELETE FROM Librarybooks where Booktitle = ?",(str(BOOKTITLE.get()),))
        con.commit()
        con.close()
        txt_result.config(text='Data deleted !!!!!!',fg='red')    
        
def Exit():
    result=mb.askquestion('Do you want to exit ? (y/n)',icon='warning')
    if result == 'yes' :
        root.destroy()
        exit()
    
        
# =========================   VARIABLE DECLARATION   ==================
        
BOOKTITLE=StringVar()
AUTHOR=StringVar()
YEAR=StringVar()
ISBN=StringVar()


# ======================  FRAME  =======================

Top=Frame(root,width=900,height=50,bd=8,relief='raise')
Top.pack(side=TOP)

Left=Frame(root,width=300,height=500,bd=8,relief='raise')
Left.pack(side=LEFT)

Right=Frame(root,width=600,height=500,bd=8,relief='raise')
Right.pack(side=RIGHT)

Forms=Frame(root,width=300,height=450)
Forms.pack(side=TOP)

Buttons=Frame(root,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)


# ====================  LABELS  ==============================

title=Label(Top,width=900,text='..... Library Management Application .....',font=('times new roman',16),bd=15)
title.pack()

booktitle=Label(Forms,text='Booktitle',font=('arial',16),bd=15)
booktitle.grid(row=0,stick='e')

author=Label(Forms,text='Author',font=('arial',16),bd=15)
author.grid(row=1,stick='e')

year=Label(Forms,text='Year',font=('arial',16),bd=15)
year.grid(row=2,stick='e')

isbn=Label(Forms,text='ISBN',font=('arial',16),bd=15)
isbn.grid(row=3,stick='e')

txt_result=Label(Buttons)
txt_result.pack(side=TOP)


# =====================   ENTRY   ============================

booktitle=Entry(Forms, textvariable=BOOKTITLE,width=30)
booktitle.grid(row=0,column=1)

author=Entry(Forms, textvariable=AUTHOR,width=30)
author.grid(row=1,column=1)

year=Entry(Forms, textvariable=YEAR,width=30)
year.grid(row=2,column=1)

isbn=Entry(Forms, textvariable=ISBN,width=30)
isbn.grid(row=3,column=1)


# =======================   BUTTONS   ============================

display=Button(Buttons,width=10,text='Display',command=Read)
display.pack(side=LEFT)

search=Button(Buttons,width=10,text='Search',command=Search)
search.pack(side=LEFT)

add=Button(Buttons,width=10,text='Add',command=Create)
add.pack(side=LEFT)

issue=Button(Buttons,width=10,text='Issue')
issue.pack(side=LEFT)

t_delete=Button(Buttons,width=10,text='Delete',command=Delete)
t_delete.pack(side=LEFT)

exit=Button(Buttons,width=10,text='Exit',command=Exit)
exit.pack(side=LEFT)


#   =========================  LIST ====================

scrollbary=Scrollbar(Right, orient=VERTICAL)
scrollbarx=Scrollbar(Right, orient=HORIZONTAL)

tree=ttk.Treeview(Right,columns=('Booktitle','Author','Year','ISBN'))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT, fill=Y)

scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=RIGHT, fill=X)

tree.heading('Booktitle',text='Booktitle',anchor=W)
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('ISBN',text='ISBN',anchor=W)

tree.column('#0', stretch=NO, minwidth=0, width=0)
tree.column('#1', stretch=NO, minwidth=0, width=80)
tree.column('#2', stretch=NO, minwidth=0, width=120)
tree.column('#3', stretch=NO, minwidth=0, width=80)

tree.pack()


#  ===========================  INITIALIZATION  =========================


if __name__ == '__main__' :
    root.mainloop()
```
