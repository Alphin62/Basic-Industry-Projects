```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('auto.csv', header=None)
data.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
data.columns = headers
data.head()
```

```python
data = data.replace('?', np.NaN)
data_1 = data.copy()          # copy of original data for further purpose
data
```

```python
data.isna().sum()
```

```python
data[['normalized-losses','num-of-doors','bore','stroke','horsepower','peak-rpm','price']].describe()
```

```python
print('-------- Average of column contains NaN values --------\n')
avg_normalized_losses = data['normalized-losses'].astype('float').mean(axis = 0)
print('normalized-losses : ', avg_normalized_losses)
data['normalized-losses'].replace(np.NaN, avg_normalized_losses, inplace = True)

data['num-of-doors'].replace(np.NaN, 'four', inplace = True)

avg_bore = data['bore'].astype('float').mean(axis = 0)
print('bore : ', avg_bore)
data['bore'].replace(np.NaN, avg_bore, inplace = True)

avg_stroke = data['stroke'].astype('float').mean(axis = 0)
print('stroke : ', avg_stroke)
data['stroke'].replace(np.NaN, avg_stroke, inplace = True)

avg_horsepower = data['horsepower'].astype('float').mean(axis = 0)
print('horsepower : ', avg_horsepower)
data['horsepower'].replace(np.NaN, avg_horsepower, inplace = True)

avg_peak_rpm = data['peak-rpm'].astype('float').mean(axis = 0)
print('peak-rpm : ', avg_peak_rpm)
data['peak-rpm'].replace(np.NaN, avg_peak_rpm, inplace = True)

avg_price = data['price'].astype('float').mean(axis = 0)
print('price : ', avg_price)
data['price'].replace(np.NaN, avg_price, inplace = True)
```
